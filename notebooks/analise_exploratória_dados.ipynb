```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
caminho = '../data/raw/base_credito_simulada_10000_enriquecida.csv'
```

```python
df = pd.read_csv(caminho)
```

```python
print(df.head())
```

```python
df.info()
```

```python
df.describe()
```

```python
# valores faltantes em cada coluna
print(df.isnull().sum())
```

```python
# Quantidade de linhas duplicadas
print("Duplicados:", df.duplicated().sum())
```

```python
# para visualizar a variável idade_cliente
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['idade_cliente'])
plt.title("Boxplot da Idade dos Clientes")
plt.show()
```

```python
# para visualizar a variável renda dos clientes
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['renda_cliente'])
plt.title("Boxplot da renda dos clientes")
plt.show()
```

```python
# para visualizar a variável Score de crédito
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['score_credito'])
plt.title("Boxplot da renda do Score de crédito")
plt.show()
```

```python
# para visualizar a variável Valor contratado
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['valor_contratado'])
plt.title("Boxplot da renda do Valor contratado")
plt.show()
```

```python
# para visualizar a variável prazo de meses
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['prazo_meses'])
plt.title("Boxplot da renda do prazo de meses")
plt.show()
```

```python
# Calcula a taxa de default por produto
default_por_produto = df.groupby('produto')['default'].mean().reset_index()
```

```python
# Visualização com gráfico de barras
plt.figure(figsize=(10,6))
sns.barplot(data=default_por_produto, x='produto', y='default')
plt.title("Taxa de Default por Produto")
plt.ylabel("Taxa de Default")
plt.xlabel("Produto")
plt.show()
```

```python
# Calcula a taxa de default por região
default_por_regiao = df.groupby('regiao')['default'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=default_por_regiao, x='regiao', y='default')
plt.title("Taxa de Default por Região")
plt.ylabel("Taxa de Default")
plt.xlabel("Região")
plt.show()
```

```python
# Cria faixas de score de crédito
bins_score = [300, 500, 700, 850]
labels_score = ['Baixo', 'Médio', 'Alto']
df['score_bin'] = pd.cut(df['score_credito'], bins=bins_score, labels=labels_score, include_lowest=True)

# Calcula a taxa de default por faixa de score
default_por_score = df.groupby('score_bin')['default'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=default_por_score, x='score_bin', y='default')
plt.title("Taxa de Default por Faixa de Score de Crédito")
plt.ylabel("Taxa de Default")
plt.xlabel("Faixa de Score")
plt.show()
```

```python
# Cria faixas de renda
bins_renda = [0, 7000, 10000, 15000, 20000]  # ajuste os limites conforme o contexto
labels_renda = ['Baixa', 'Média-Baixa', 'Média-Alta', 'Alta']
df['renda_bin'] = pd.cut(df['renda_cliente'], bins=bins_renda, labels=labels_renda, include_lowest=True)

# Calcula a taxa de default por faixa de renda
default_por_renda = df.groupby('renda_bin')['default'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=default_por_renda, x='renda_bin', y='default')
plt.title("Taxa de Default por Faixa de Renda")
plt.ylabel("Taxa de Default")
plt.xlabel("Faixa de Renda")
plt.show()
```
